```python
from application_logging.logger import App_Logger
import pandas as pd
import os
```

```python
os.mkdir('logFiles')
logger  = App_Logger('logFiles/getFilePath.log')
logger1 = App_Logger('logFiles/readFilePath.log')
logger2 = App_Logger('logFiles/data_Combine.log')
```

```python
class dataCombine:
    
    def __init__(self,path):
        self.path = path
        
    def getFilePath(self):
        try:
            logger.info('INFO','Getting The File Path of Each File')
            file_path = []
            for root, directory, file in os.walk(self.path):
                for filename in file:
                    get_file_path=os.path.join(root,filename)
                    file_path.append(get_file_path)
            logger.info('INFO', 'Getting File Path As Follows : \n {}'.format(file_path))
            return file_path
        
        except Exception as e:
            raise Exception(f"(getFilePath)-Something Went Wrong With Fucntion \n" + str(e))
            
    def readFilePath(self):
        try:
            logger1.info('INFO','Reading The File Path of Each File')
            main = []
            file = self.getFilePath()
            for i in file:
                df = pd.read_table(i,sep=',',header=None)
                logger1.info('INFO','Reading The File Path No. {Number} And Shape Is {Shape}'.format(Number=i.split("\\")[-1],
                                                                                                     Shape=df.shape))
                main.append(df)
            return main
        
        except Exception as e:
            raise Exception(f"(readFilePath)-Something Went Wrong With Fucntion \n" + str(e))
            
    def combineFile(self):
        try:
            logger2.info('INFO','Combine All The File')
            data = self.readFilePath()
            df = pd.concat(data)
            new_data = df.rename(columns={0:'Time',1:'Acceleration_frontal',2:'Acceleration_vertical',3:'Acceleration_lateral',
                                          4:'Antenna',5:'RSSI',6:'Phase',7:'Frequency',8:'Label'})
            new_data.to_csv('Datasets_Healthy_Older_People/Combine.csv')
            logger2.info('INFO','Combined All Files Having Record {}'.format(new_data.shape))
            return new_data
        
        except Exception as e:
            raise Exception(f"(combineFile)-Something Went Wrong With Fucntion \n" + str(e))
```

```python
d = dataCombine('Datasets_Healthy_Older_People')
```

```python
d.combineFile()
```
